##### AIM:
Write a program to plot sin/cosine signal, step signal, ramp signal, parabola, exponential, impulse and rectangular signal in the continuous and discrete form.

Python

##### THEORY:
Real signals can be quite complicated. The study of signals therefore starts with the analysis of basic and fundamental signals. For linear systems, a complicated signal and its behaviour can be studied by superposition of basic signals. Common basic signals are sin/cosine signal, step signal, ramp signal, parabola, exponential, impulse and rectangular signal.


##### CODE:

```python
import numpy as np
import matplotlib.pyplot as plt

#Sine Signal:
f=float(input ("Enter the frequency:"))
t=np.arange(0,1,0.01)  #x-axis
y=np.sin(2*f*t*np.pi) #y-axis
#continuos plot
plt.plot(t,y)
plt.title("Continuos plot-sin wave")
plt.show()
#discrete plot
plt.stem(t,y)
plt.title("Discrete plot-sin wave")
plt.show()
```

```python
#Step Signal:
l,u=input("Enter the lower and upper range of the function").split(",",2)
t=np.arange(int(l),int(u)+1,1)  #x-axis; u+1 includes the upper range limit
y=np.zeros(len(t))              #y-axis
for i in range (0,len(t)):
  if t[i]<=0:
    y[i]=0
  else:
    y[i]=1

#continuos plot
plt.plot(t,y)
plt.title("Continuos plot-step signal")
plt.show()
#discrete plot
plt.stem(t,y)
plt.title("Discrete plot-step signal")
plt.show()
```

```python
#Ramp Signal:
l,u=input("Enter the lower and upper range of the function:").split(",")
t=np.arange(int(l),int(u)+1,1)  #x-axis; u+1 includes the upper range limit
y=np.zeros(len(t))              #y-axis
for i in range (0,len(t)):
  if t[i]<=0:
    y[i]=0
  else:
    y[i]=t[i]
#continuos plot
plt.plot(t,y)
plt.title("Continuos plot-ramp signal")
plt.show()
#discrete plot
plt.stem(t,y)
plt.title("Discrete plot-ramp signal")
plt.show()
```

```python
#Parabolic Signal:
l,u=input("Enter the lower and upper range of the function:").split(",",2)
t=np.linspace(int(l),int(u),100) #x-axis;linspace includes both upper and lower limit values
a = float(input("Enter the value of coefficent a :"))
y = a*(t**2)                     #y-axis
#continuos plot
plt.plot(t,y)
plt.title("Continuos plot-parabolic signal")
plt.show()
#discrete plot
plt.stem(t,y)
plt.title("Discrete plot-parabolic signal")
plt.show()
```

```python
#Exponential Signal:
l,u=input("Enter the lower and upper range of the function:").split(",",2)
t=np.arange(int(l),int(u)+1,0.1)  #x-axis; u+1 includes the upper range limit
y = np.exp(t)
#continuos plot
plt.plot(t,y)
plt.title("Continuos plot-Exponential signal")
plt.show()
#discrete plot
plt.stem(t,y)
plt.title("Discrete plot-Exponential signal")
plt.show()
```

```python
#Impulse Signal:
l,u=input("Enter the lower and upper range of the function").split(",",2)
t=np.arange(int(l),int(u)+1)  #x-axis; u+1 includes the upper range limit
x=np.zeros(len(t))
for i in range (len(t)):
  if t[i]!=0:
    x[i]=0
  else:
    x[i]=1
#continuos plot
plt.plot(t,x)
plt.title("Continuos plot-Impulse signal")
plt.show()
#discrete plot
plt.stem(t,x)
plt.title("Discrete plot-Impulse signal")
plt.show()

```

```python
#Rectangular signal:
w = int(input("Enter the width of the pulse :"))
e = float(w/2)
n = -1*e
print(e,n)
t = np.arange(n,e+0.1,1)
bt= np.arange(-20,n,0.1)
at= np.arange(e+0.1,20.1,1)
x = np.concatenate((bt,t,at))
bv= np.zeros(len(bt))
av= np.zeros(len(at))
v = np.ones(len(t))
y = np.concatenate((bv,v,av))
#continuos plot
plt.plot(x,y)
plt.title("Continuous plot-Impulse signal")
plt.show()
#discrete plot
plt.stem(x,y)
plt.title("Discrete plot-Impulse signal")
plt.show()
```
