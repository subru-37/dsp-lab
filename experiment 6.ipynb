##### CODE:

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
x = np.array(input("Enter the elements of the input sequence:").split(",")).astype(complex)
k = len(x)
y = np.zeros(k,dtype = 'complex')

for i in range (k):
  sum = 0
  for m in range (k):
    sum += x[m]*np.exp(complex(1j)*2*np.pi*i*m*float(1/k))
  y[i] = sum/k
print("The IDFT of the input sequence using Mathematical expression is:\n",y)
k_array = np.arange(0,k,1)
mag_y = [abs(k) for k in y ]
plt.stem(k_array,mag_y)
plt.title("Magnitude Plot")
plt.xlabel("Value of k")
plt.ylabel("|X[k]|")
plt.show()

#Phase Plot
phase_y = [cmath.phase(k) for k in y]
plt.stem(k_array,phase_y)
plt.title("Phase Plot")
plt.xlabel("Value of k")
plt.ylabel("<X[k]")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array(input("Enter the elements of the input sequence:").split(",")).astype(complex)
k = len(x)

twiddle = []
k_array = np.arange(0,k,1)
for i in range(k):
    temp = []
    for j in range(k):
        if i==0:
            temp.append(complex(1))
        elif j==0:
            temp.append(complex(1))
        else:
            temp.append(np.exp(complex(1j)*2*np.pi*i*j*float(1/k)))
    twiddle.append(temp)
# print(twiddle)
y=np.dot(x,twiddle)/k
print("The IDFT of the input sequence using Matrix method is:\n",np.dot(x,twiddle)/k)

mag_y = [abs(k) for k in y ]
plt.stem(k_array,mag_y)
plt.title("Magnitude Plot")
plt.xlabel("Value of k")
plt.ylabel("|X[k]|")
plt.show()

#Phase Plot
phase_y = [cmath.phase(k) for k in y]
plt.stem(k_array,phase_y)
plt.title("Phase Plot")
plt.xlabel("Value of k")
plt.ylabel("<X[k]")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array(input("Enter the elements of the input sequence:").split(",")).astype(complex)
N = int(input('enter N'))
if len(x)<N:
    x = np.pad(x,(0,N-len(x)))
else:
    print('error')
k = len(x)
twiddle = []

for i in range(k):
    temp = []
    for j in range(k):
        if i==0:
            temp.append(complex(1))
        elif j==0:
            temp.append(complex(1))
        else:
            temp.append(np.exp(complex(1j)*2*np.pi*i*j*float(1/k)))
    twiddle.append(temp)
# print(twiddle)
y = np.dot(x,twiddle)/k
print("The IDFT of the input sequence using Matrix method is:\n",np.dot(x,twiddle)/k)
k_array = np.arange(0,len(y),1)
mag_y = [abs(k) for k in y ]
plt.stem(k_array,mag_y)
plt.title("Magnitude Plot")
plt.xlabel("Value of k")
plt.ylabel("|X[k]|")
plt.show()

#Phase Plot
phase_y = [cmath.phase(k) for k in y]
plt.stem(k_array,phase_y)
plt.title("Phase Plot")
plt.xlabel("Value of k")
plt.ylabel("<X[k]")
plt.show()
```
